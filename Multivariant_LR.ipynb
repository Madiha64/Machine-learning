```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("D:\\madiha study\\data\\homeprice.csv")
df.head()
```

```python
 # the way we will take the median if there is no value in the dataset to solve that matter
df.bedrooms.median()
```

```python
# to remove the fraction part if the median is 4.5 etc then we code like follow:
import math
median_bedrooms=math.floor(df.bedrooms.median())
median_bedrooms
```

```python
# then put this median number into data like:
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df.drop("price",axis="columns"),df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old
reg.predict([[2500,4,5]])
```
