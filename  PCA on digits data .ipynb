```python
# let me create a dataframe now.
df=pd.DataFrame(digits.data)
df.head()
```

```python
df.describe()
```

```python
X=df
y=digits.target
```

```python
X  # Now X is our data
```

```python
y # y is our class or output
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca
```

```python
pca=PCA(0.95) # it mean i need the data 95% usefull
x_pca_scaled=pca.fit_transform(X)
x_pca_scaled
```

```python
x_pca_scaled.shape # here is columns reductions 
```

```python
# in old datasets we have 
X.shape  # number of columns reduced due to PCA.It is used for dimensiuonallity reduction
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
# when this pca data we used for training than 
X_train_pca,X_test_pca,y_train_pca,y_test_pca=train_test_split(x_pca_scaled,y,test_size=0.2,random_state=30)
model= LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)
```

```python
# now if we want the components in more reduced form
pca=PCA(n_components=2)
x_pca1=pca.fit_transform(X)
x_pca1.shape
```

```python
x_pca1 # in the output we have only two columns
```

```python
x_pca1.explained_variance_ratio_
```

```python
X_train_pca,X_test_pca,y_train_pca,y_test_pca=train_test_split(x_pca1,y,test_size=0.2,random_state=30)
model= LogisticRegression(max_iter=1000)
model.fit(X_train_pca,y_train_pca)
model.score(X_test_pca,y_test_pca)
# now in the output we looking the accuracy is low due to reducing the n_components it will decrease the computation power but not increase the accuracy.
# in this way lot of data will be missed.
```
